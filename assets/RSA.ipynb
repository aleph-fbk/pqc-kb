```python
def order(b,N):
    f=[]
    for r in range(2,N):
        f.append( pow(b,r,N) )    
    try:
        order = f.index(1)+2
        return order
    except:
        return '_'
```

```python
def period(b,N):
    f=[]
    for r in range(2,N):
        f.append( pow(b,r,N) )    
    try:
        period = f.index(b)+1
        print(f'{b}: {f}')
        return period
    except:
        return '_'
```

```python
N = 13*17 # modulus
#b = 2 # base
o = []
p = []
for b in range(2,N-1):
    o.append( order(b,N) )
    p.append( period(b,N) )
print(f'orders: {o}')
print(f'periods: {p}')
```
