## Order finding

```python
def order(b,N):
    f=[]
    for r in range(2,N):
        f.append( pow(b,r,N) )    
    try:
        order = f.index(1)+2
        return order
    except:
        return '_'
```

```python
def period(b,N):
    f=[]
    for r in range(2,N):
        f.append( pow(b,r,N) )    
    try:
        period = f.index(b)+1
        print(f'{b}: {f}')
        return period
    except:
        return '_'
```

```python
N = 13*17 # modulus
#b = 2 # base
o = []
p = []
for b in range(2,N-1):
    o.append( order(b,N) )
    p.append( period(b,N) )
print(f'orders: {o}')
print(f'periods: {p}')
```

```python
from math import gcd
gp = []
gm = []

for b in range(2,N-1):
    b_squared = pow(b,2,N)
    gm.append( gcd(b-1,N) )
    gp.append( gcd(b+1,N) )
print(gm)
print(gp)
```

```python
gcdp = []

for j in range(len(p)):
    gcdp.append( gcd(p[j]+1,N) )
print(gcdp)
```

## Scaling

```python
import numpy

# key length in bits -- n = log N
n = numpy.arange(128, 4097, 16, dtype=int)
n_idx = [numpy.argwhere(n==x)[0][0] for x in [1024,2048,3072,4096]]
#B_star=numpy.array([1024,2048,3072,4096], dtype=int)

GNFS = numpy.exp(1.923*(n)**(1/3)*(numpy.log(n))**(2/3))
Shor = (n**3)#*numpy.log(n)
KMYR = (n**1.8)
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
#ax.axhline(y=0, color="black", linestyle="--")
#ax.axhline(y=0.5, color="black", linestyle=":")
#ax.axhline(y=1.0, color="black", linestyle="--")
#ax.axvline(color="grey", x=1024)
#ax.axline((0, 0.5), slope=0.25, color="black", linestyle=(0, (5, 5)))

# GNFS (ref: logjam)
ax.semilogy(n, GNFS, linewidth=2, linestyle="--", label=r'GNFS: $\approx e^{n^{1/3} \log(n)^{2/3}}$', base=2)
ax.semilogy(n[n_idx], GNFS[n_idx], linewidth=0, marker='o', fillstyle='none', base=2)

# Shor (ref: BSI)
ax.semilogy(n, Shor, linewidth=2, label=r'Shor: $\approx n^3$', base=2)
ax.semilogy(n[n_idx], Shor[n_idx], linewidth=0, marker='o', fillstyle='none', base=2)

# Regev w/ Kahanamoku-Meyer_Yao_24
ax.semilogy(n, KMYR, linewidth=2, linestyle=":", label=r'Regev + KMY: $\approx n^{1.8}$', base=2)
ax.semilogy(n[n_idx], KMYR[n_idx], linewidth=0, marker='o', fillstyle='none', base=2)


ax.set(xlabel='n = log N -- RSA modulus bit length', ylabel=r'complexity to factor / $\approx$ bits of security')
ax.legend(fontsize=14, framealpha=1)
plt.show()
fig.savefig('temp.png', transparent=True)
```
